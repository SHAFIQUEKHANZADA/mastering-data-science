# **Phase 3 - Practice With Real-World Data**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 12000, 13500, 14200, 15000],
    [2, 9500, 11000, 11750, 12500],
    [3, 13400, 13900, 14300, 14900],
    [4, 8800, 9400, 10100, 11000],
    [5, 15000, 15800, 16200, 17000]
])

print("==== Sales Analysis ====")
print("\n Sales Data Shape:", sales_data.shape)
print("\n Sample Data for 1st 3 Restaurant:")
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

# minimum sales per restaurant
min_data = np.min(sales_data[:, 1:], axis=1)
print(min_data)

# maximum sales per year
max_data = np.max(sales_data[:, 1:], axis=0)
print(max_data)

# average sales per year
avg_data = np.mean(sales_data[:, 1:], axis=0)
print(avg_data)
```

## **Cumulative Sum in NumPy**
The cumulative sum (or cumsum) means each element in the output is the sum of all previous elements up to that point.

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative Sales Accross All Restaurants")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# **Operations of Vectors**

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

# Addition
result_add = vector1 + vector2
print("Addition:", result_add)

# Subtraction
result_sub = vector1 - vector2
print("Subtraction:", result_sub)

# Multiplication
print("Multiplication:", vector1 * vector2)

# Division
print("Division:", vector1 / vector2)

# Dot product
print("Dot Product:", np.dot(vector1, vector2))

# Vectorized operations
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle:", np.degrees(angle))
```
