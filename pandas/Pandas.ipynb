# **Series**

One-dimensional data structure

```python
import pandas as pd

var = [10, 20, 30, 40]
series = pd.Series(var, index=["a", "b", "c", "d"], dtype="float", name="First Data")
print(series)
print(series["d"])
print(type(series))
```

```python
dic = {
    "name" : ["Shafique", "Zayan", "Hamdan"],
    "age" : [22, 21, 19],
    "hobbies" : ["Chess", "Football", "Nothing"]
}

x = pd.Series(dic)
x
```

```python
x1 = pd.Series(12, index=[1,2,3,4,5])
x2 = pd.Series(12, index=[1,2,3])
print(x1 + x2)
```

# **DataFrame**

Two-dimensional data structure

```python
l = [1,2,3,4,5]

var = pd.DataFrame(l)
print(type(var))
print(var)
```

```python
dicx = {
    "a" : [1,2,3,4,5],
    "b" : [6,7,8,9,10],
    "c" : [11,12,13,14,15],
    "d" : [16,17,18,19,20]
}

var1 = pd.DataFrame(dicx, columns=["a", "d"], index=["A", "B", "C", "D", "E"])
print(type(var1))
print(var1)
print(var1["a"][3])
```

```python
lists = [[1,2,3], [4,5,6], [7,8,9]]

var2 = pd.DataFrame(lists, columns=["A", "B", "C"])
print(var2)
```

```python
dis_ser = {
    "x" : pd.Series([10,20,30]),
    "y" : pd.Series([11,21,31]),
    "z" : pd.Series([12,22,32]),
}

var3 = pd.DataFrame(dis_ser)
print(type(var3))
print(var3)
```

# **Arthmatic Operator in Pandas**

# **Merge**

```python
import pandas as pd

data = pd.DataFrame({
    "a" : [1, 2, 3, 4, 5],
    "b" : [11, 12, 13, 14, 15]
})
data1 = pd.DataFrame({
    "a" : [1, 2, 3, 4, 5],
    "c" : [21, 22, 23, 24, 25]
})

pd.merge(data, data1, on="a")
```

# **Concat**

```python
x = pd.Series([1,2,3,4,5])
y = pd.Series([6,7,8,9,10])

pd.concat([x,y])
```

```python
a = pd.DataFrame({
    "a" : [1, 2, 3, 4, 5],
    "b" : [11, 12, 13, 14, 15]
})
b = pd.DataFrame({
    "a" : [1, 2, 3, 4, 5],
    "b" : [21, 22, 23, 24, 25]
})

pd.concat([a,b])
```

# **Join**

```python
j1 = pd.DataFrame({"names": ["Shafique", "Zayan", "Hamdan"], "age": [22, 21, 19]})
j2 = pd.DataFrame({"subject": ["Artificial Intelligence", "Forensic Science", "Computer Science"], "hobbies": ["Chess", "Football", "Nothing"]})

j1.join(j2)
```
